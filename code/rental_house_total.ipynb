```python
import requests
import json
import pandas as pd
import warnings

warnings.filterwarnings('ignore')

siCode = ['110', '170', '200', '215', '230', '260', '290', '305', '320', '350', '380', '410', '440', '470', '500', '530', '545', '560', '590', '620', '650', '680', '710', '740']

url = 'https://data.myhome.go.kr/rentalHouseList?serviceKey=SIMuYtPaPVHuKqIUXGQXnLCDnSXMWsyauPFvNDsI10qfpdQ7po0WKOaDryWxfEaseFwFbdJgmTIqydTZr31Wlw%3D%3D&brtcCode=11&si'

all_data = []

for code in siCode:
    params = {'signguCode': code, 'numOfRows': '100', 'pageNo': '1'}

    try:
        response = requests.get(url, params=params, verify=False)
        contents = response.text
        json_ob = json.loads(contents)
        data = json_ob['hsmpList']

        for item in data:
            extracted_item = {
                'brtcNm': item['brtcNm'],
                'signguNm': item['signguNm'],
                'hsmpNm': item['hsmpNm'],
                'competDe': item['competDe'],
                'hshldCo': item['hshldCo'],
                'suplyTyNm': item['suplyTyNm'],
                'suplyPrvuseAr': item['suplyPrvuseAr'],
                'houseTyNm': item['houseTyNm'],
                'buldStleNm': item['buldStleNm'],
                'elvtrInstlAtNm': item['elvtrInstlAtNm'],
                'bassRentGtn': item['bassRentGtn'],
                'bassMtRntchrg': item['bassMtRntchrg'],
                'bassCnvrsGtnLmt': item['bassCnvrsGtnLmt']
            }
            all_data.append(extracted_item)

    except Exception as e:
        print(f"Error fetching data for signguCode {code}: {e}")

# Create DataFrame
seoul_df = pd.DataFrame(all_data)
seoul_df.columns = ['광역시도', '시군구', '단지', '준공 일자', '세대 수', '공급 유형',
              '공급 전용 면적', '주택 유형', '건물 형태', '승강기 설치여부',
              '기본 임대보증금', '기보 월임대료', '기본 전환보증금']

# Show the DataFrame
print(seoul_df)
```

```python
#한글깨짐 방지
seoul_df.to_csv('rentalhouse_seoul.csv',index=False, encoding='utf-8-sig')
```

```python
import requests
import json
import pandas as pd
import warnings

warnings.filterwarnings('ignore')

siCode = ['110', '140', '170', '200', '215', '230', '260', '290', '305', '320', '350', '380', '410', '440', '470', '500', '530', '545', '560', '590', '620', '650', '680', '710', '740',
          '185', '237', '245', '720', '155', '111', '113',
          '115', '117', '130', '131', '133', '135', '150', '171', '173', '190', '210',
          '220', '250', '270', '271', '273', '280', '281', '285', '287', '310', '360',
          '370', '390', '430', '450', '460', '461', '463', '465', '480', '550', '570',
          '610', '630', '670', '800', '820', '830', '730', '750', '760', '770', '780',
          '790', '810', '112', '114', '745', '180', '825', '840', '860', '870', '880',
          '890', '900', '910', '850', '920', '930', '940', '120', '121', '123', '125',
          '127', '129', '240', '330']
brtcCode = ['11', '26', '27', '28', '29', '30', '31', '41', '42', '43', '44', '45', '46', '47', '48', '50']

url = 'https://data.myhome.go.kr/rentalHouseList?serviceKey=SIMuYtPaPVHuKqIUXGQXnLCDnSXMWsyauPFvNDsI10qfpdQ7po0WKOaDryWxfEaseFwFbdJgmTIqydTZr31Wlw%3D%3D'

all_data = []

for b_code in brtcCode:
    for s_code in siCode:
        params = {'brtcCode': b_code, 'signguCode': s_code, 'numOfRows': '100', 'pageNo': '1'}

        try:
            response = requests.get(url, params=params, verify=False)
            contents = response.text
            json_ob = json.loads(contents)

            # Check if 'hsmpList' key exists and has data
            if 'hsmpList' in json_ob and json_ob['hsmpList']:
                data = json_ob['hsmpList']

                for item in data:
                    extracted_item = {
                        'brtcNm': item['brtcNm'],
                        'signguNm': item['signguNm'],
                        'hsmpNm': item['hsmpNm'],
                        'competDe': item['competDe'],
                        'hshldCo': item['hshldCo'],
                        'suplyTyNm': item['suplyTyNm'],
                        'suplyPrvuseAr': item['suplyPrvuseAr'],
                        'houseTyNm': item['houseTyNm'],
                        'buldStleNm': item['buldStleNm'],
                        'elvtrInstlAtNm': item['elvtrInstlAtNm'],
                        'bassRentGtn': item['bassRentGtn'],
                        'bassMtRntchrg': item['bassMtRntchrg'],
                        'bassCnvrsGtnLmt': item['bassCnvrsGtnLmt']
                    }
                    all_data.append(extracted_item)
            else:
                print(f"No data for brtcCode {b_code} and signguCode {s_code}")

        except Exception as e:
            print(f"Error fetching data for brtcCode {b_code} and signguCode {s_code}: {e}")

# Create DataFrame
df = pd.DataFrame(all_data)
df.columns = ['광역시도', '시군구', '단지', '준공 일자', '세대 수', '공급 유형',
              '공급 전용 면적', '주택 유형', '건물 형태', '승강기 설치여부',
              '기본 임대보증금', '기본 월임대료', '기본 전환보증금']

# Show the DataFrame
print(df)
```

```python
# Save DataFrame to CSV with UTF-8 encoding
df.to_csv('rentalhouse_total.csv', index=False, encoding='utf-8-sig')
```
