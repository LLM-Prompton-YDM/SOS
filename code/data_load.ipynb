```python
# request 요청
response = requests.get(url)
contents = response.text
```

```python
# 데이터 출력 및 확인
pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents))
```

```python
# json 형태로 출력
json_ob = json.loads(contents)
print(json_ob)
print(type(json_ob))
```

```python
# 청약 정보 추출
# PAN_NT_ST_DT : 공고게시일
# PAN_ID : 공고아이디
# AIS_TP_CD_NM : 공고세부유형명
# CNP_CD_NM : 지역명
# ALL_CNT : 전체조회건수
# SPL_INF_TP_CD : 공급정보구분코드
# AIS_TP_CD : 매물유형코드
# PAN_DT : 모집공고일
# RNUM : 순번
# CCR_CNNT_SYS_DS_CD : 고객센터연계시스템구분코드
# DTL_URL : 공고상세 URL
# CLSG_DT : 공고마감일
# UPP_AIS_TP_CD : 공급정보구분코드
# PAN_NM : 공고명
# UPP_AIS_TP_NM : 공고유형명
# PAN_SS : 공고상태
# DTL_URL_MOB : 공고상세 URL(DTL_URL과 동일)
data = json_ob[1]['dsList']
data[0]
```

```python
# 필요한 정보만 추출
# 공고게시일, 공고세부유형명, 지역명, 공고상태, 공고유형명, 공고명, 공고상세 URL, 공고마감일
features = ['PAN_NT_ST_DT', 'AIS_TP_CD_NM', 'CNP_CD_NM', 'PAN_SS', 'UPP_AIS_TP_NM', 'PAN_NM', 'DTL_URL', 'CLSG_DT']

extracted_data = []

for item in data:
    extracted_item = {key: item[key] for key in features}
    extracted_data.append(extracted_item)

for item in extracted_data:
    print(item)
```
