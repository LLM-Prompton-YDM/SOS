```python
# brtcNm : 광역시도 명
# signguNm : 시군구  명
# hsmpNm : 단지 명
# competDe : 준공 일자
# hshldCo : 세대 수
# suplyTyNm : 공급 유형 명
# suplyPrvuseAr : 공급 전용 면적
# houseTyNm : 주택 유형 명
# buldStleNm : 건물 형태
# elvtrInstlAtNm : 승강기 설치여부
# bassRentGtn : 기본 임대보증금
# bassMtRntchrg : 기본 월임대료
# bassCnvrsGtnLmt : 기본 전환보증금
data = json_ob['hsmpList']
```

```python
features = ['brtcNm', 
            'signguNm', 
            'hsmpNm', 
            'competDe',
            'hshldCo',
            'suplyTyNm',
            'suplyPrvuseAr',
            'houseTyNm',
            'buldStleNm',
            'elvtrInstlAtNm',
            'bassRentGtn',
            'bassMtRntchrg',
            'bassCnvrsGtnLmt']

extracted_data = []

for item in data:
    extracted_item = {key: item[key] for key in features}
    extracted_data.append(extracted_item)

for item in extracted_data:
    print(item)
```
